```python
# Step 2: Import necessary libraries and load the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Load the data
df = pd.read_csv("survey_data_updated 5.csv")

# Display the first few rows to understand the structure of the data
df.head()

# Ensure unique respondent ID exists
if "ResponseId" not in df.columns:
    df["ResponseId"] = df.index

# Columns to split
split_columns = [
    "LanguageHaveWorkedWith",
    "DatabaseHaveWorkedWith",
    "PlatformHaveWorkedWith",
    "WebframeHaveWorkedWith"
]

# Function to split and explode columns
def split_and_explode(df, column):
    return (
        df[["ResponseId", column]]
        .dropna()
        .assign(**{column: df[column].str.split(";")})
        .explode(column)
        .rename(columns={column: "Value"})
        .assign(SourceColumn=column)
        .reset_index(drop=True)
    )

# Apply splitting
split_dfs = [split_and_explode(df, col) for col in split_columns]

# Combine all split data
split_only_df = pd.concat(split_dfs, ignore_index=True)

# Clean whitespace
split_only_df["Value"] = split_only_df["Value"].str.strip()

# ✅ Sort by ResponseId (primary) and SourceColumn (secondary – optional)
split_only_df = split_only_df.sort_values(
    by=["ResponseId", "SourceColumn", "Value"]
).reset_index(drop=True)

# Save to separate CSV file
split_only_df.to_csv(
    "survey_data_techstack_worked_with.csv",
    index=False
)

print("✅ Split-only CSV sorted by ResponseId and saved successfully")

split_only_df.head(20)
```

```python
import pandas as pd

# Load original dataset
df = pd.read_csv("survey_data_updated 5.csv")

# Ensure unique respondent ID exists
if "ResponseId" not in df.columns:
    df["ResponseId"] = df.index

# Mapping of columns to output filenames
split_config = {
    "LanguageHaveWorkedWith": "survey_languages_workwith.csv",
    "DatabaseHaveWorkedWith": "survey_databases_workwith.csv",
    "PlatformHaveWorkedWith": "survey_platforms_workwith.csv",
    "WebframeHaveWorkedWith": "survey_webframes_workwoth.csv"
}

def split_save_csv(df, column, out_file):
    split_df = (
        df[["ResponseId", column]]
        .dropna()
        .assign(**{column: df[column].str.split(";")})
        .explode(column)
        .rename(columns={column: "Value"})
        .reset_index(drop=True)
    )

    split_df["Value"] = split_df["Value"].str.strip()

    # Sort by ResponseId and Value
    split_df = split_df.sort_values(
        by=["ResponseId", "Value"]
    ).reset_index(drop=True)

    # Save CSV
    split_df.to_csv(out_file, index=False)
    print(f"✅ Saved {out_file}")

# Run for each column
for col, file in split_config.items():
    split_save_csv(df, col, file)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("survey_data_updated 5.csv")

# Columns and output mapping
split_config = {
    "LanguageHaveWorkedWith": "top10_languages_workedwith.csv",
    "DatabaseHaveWorkedWith": "top10_databases_workedwith.csv",
    "PlatformHaveWorkedWith": "top10_platforms_workedwith.csv",
    "WebframeHaveWorkedWith": "top10_webframes_workedwith.csv",
    "LanguageWantToWorkWith": "top10_languages_wanttowork.csv",
    "DatabaseWantToWorkWith": "top10_databases_wanttowork.csv",
    "PlatformWantToWorkWith": "top10_platforms_wanttowork.csv",
    "WebframeWantToWorkWith": "top10_webframes_wanttowork.csv"
    
    
    
}

def save_top10_splits(df, column, out_file):
    top10 = (
        df[column]
        .dropna()
        .str.split(";")
        .explode()
        .str.strip()
        .value_counts()
        .head(10)
        .reset_index()
        .rename(columns={"index": "Value", column: "Count"})
    )

    top10.to_csv(out_file, index=False)
    print(f"✅ Saved {out_file}")

# Generate Top 10 CSVs
for col, file in split_config.items():
    save_top10_splits(df, col, file)
```
